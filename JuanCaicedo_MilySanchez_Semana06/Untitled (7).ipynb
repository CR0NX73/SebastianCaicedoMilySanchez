```python
import numpy as np
import matplotlib.pyplot as plt

tmin=0.
tmax=10.
N = 200
deltat = (tmax-tmin)/N
t = np.linspace(tmin, tmax, N)

tmin1 = tmin - 0.5*deltat
tmax1 = tmax - 0.5*deltat
t1 = np.arange(tmin1, tmax1+deltat, deltat)
x0 = 1
v0 = 0

def GetA(x):
    return -x
def GetLeapFrog(x,v,deltat):
    for i in range(1,N):
        v[i] = v[i-1] + GetA(x[i-1])*deltat
        x[i] = x[i-1] + v[i]*deltat


x = np.zeros(N)
x[0] = x0
v = np.zeros(N)
v[0] = v0 - 0.5*deltat*GetA(x[0])

GetLeapFrog(x,v,deltat)
X = np.array([])

for i in range(len(x)-1):
    X = np.append(X , 0.5*(x[i]+x[i+1]))
V = v[1:]
E = 0.5*X**2 + 0.5*V**2
plt.plot(t1[2:],E,c="g",label="Leap-Frog")


c1=0.6756
c2=-0.1756
c3=-0.1756
c4=0.6756


x=np.zeros(200)
x[0]=1
v=np.zeros(200)

w0=-(2**(1/3))/(2-2**(1/3))
w1=1/(2-2**(1/3))
d1=w1
d2=w0
d3=w1

for i in range(1,len(x)):
    x_1=x[i-1]+c1*v[i-1]*deltat
    v_1=v[i-1]+d1*GetA(x_1)*deltat
    x_2=x_1+c2*v_1*deltat
    v_2=v_1+d2*GetA(x_2)*deltat
    x_3=x_2+c3*v_2*deltat
    v_3=v_2+d3*GetA(x_3)*deltat
    x_4=x_3+c4*v_3*deltat
    
    x[i]=x_4
    v[i]=v_3

Ey=(v**2)/2 + (x**2)/2
plt.plot(t,Ey,c="r",label="Yoshida")
    
    
```
